```python
import PyPDF2
import nltk 
```

```python
file = open('Elneny.pdf', 'rb')
pdf = PyPDF2.PdfFileReader(file)
page = pdf.getPage(0)
text = page.extractText()
print(text)
```

```python
nltk.word_tokenize(text)
```

```python
ps = nltk.PorterStemmer()
```

```python
for w in text:
    print(ps.stem(w))
```
